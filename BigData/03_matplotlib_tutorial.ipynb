## 기본 그래프

```python
x = [1, 3, 5]
y = [2, 8, 11]
```

```python
plt.plot(x, y)
plt.title("Graph")
plt.xlabel("x", loc="right")
plt.ylabel("y", loc="top")
plt.show()
```

```python
plt.rcParams.get("figure.figsize")
```

#### Figure, Axes
- ![](https://storage.googleapis.com/skytowner_public/images/uG4nzrYfQ9KDk2M87YDI/matplotlib_figure_axes_axis%20(4).png)

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)   # 2행 2열 그림을 그리는데 첫번째 것을 의미, ax1 = fig.add_subplot(221)와 같음
ax2 = fig.add_subplot(2, 2, 2)   # 2행 2열 그림을 그리는데 두번째 것을 의미, ax2 = fig.add_subplot(222)와 같음
ax3 = fig.add_subplot(2, 2, 3)   # 2행 2열 그림을 그리는데 세번째 것을 의미, ax3 = fig.add_subplot(223)와 같음
ax4 = fig.add_subplot(2, 2, 3)   # 2행 2열 그림을 그리는데 네번째 것을 의미, ax4 = fig.add_subplot(224)와 같음

ax1.plot(x, y)
ax1.set_title("Graph")
ax4.plot(x, y)

plt.show()
```

```python
x = [1, 3, 5]
y = [2, 8, 11]
z = [3, 4, 5]

# plt.plot(x, y, color="r", linestyle=":")   # "g"나 "b"로 한글자만 써서 바꿀 수 있고 "red"나 "green"으로 써도 바꿀 수 있음
plt.plot(x, y, "b--", marker="^")
plt.plot(x, z, color="midnightblue", linewidth=5)
plt.show()
```

```python
x = [1, 3, 5]
y = [2, 8, 11]
z = [3, 4, 5]

plt.plot(x, y, label="Eng")
plt.plot(x, z, label="Math")
plt.legend(loc="lower right")   # plt.plot()에서 label 옵션 사용했다면 plt.legend() 써야 적용됨
plt.show()
```

## 막대 그래프(bar plot)
* 범주형 데이터를 다룰 때 사용

```python
x = ["A", "B", "C", "D", "E"]
y = [3, 2, 5, 4, 6]

plt.bar(x, y, color="r")
plt.show()
```

## 히스토그램(Histogram)
- 연속형 데이터를 다룰 때 사용
- 기본적으로 10개의 구간으로 나눔

```python
import numpy as np

mydata = np.random.randn(100)
# plt.hist(mydata)
plt.hist(mydata, bins=20)   # bins 옵션 : 구간 개수 설정
plt.show()
```

## 산점도(Scatter plot)
- 분포를 파악하는데 좋음

```python
!gdown https://raw.githubusercontent.com/devdio/datasets/main/iris.csv
```

```python
import pandas as pd

df = pd.read_csv("iris.csv", index_col=0)
df.head()
```

```python
plt.scatter(df["SepalLengthCm"], df["SepalWidthCm"], color="r")
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title("iris")
plt.show()
```

```python
print(df["Species"].unique(), "\n\n")
print(df["Species"].value_counts())
```

```python
# Species에 따라 SepalLengthCm과 SepalWidthCm을 색상을 달리해 표현

setosa = df[df["Species"] == "Iris-setosa"]
versicolor = df[df["Species"] == "Iris-versicolor"]
virginica = df[df["Species"] == "Iris-virginica"]

plt.scatter(setosa["SepalLengthCm"], setosa["SepalWidthCm"], color="r")
plt.scatter(versicolor["SepalLengthCm"], versicolor["SepalWidthCm"], color="g")
plt.scatter(virginica["SepalLengthCm"], virginica["SepalWidthCm"], color="b")

plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title("iris")
plt.show()
```

```python
# Species에 따라 SepalLengthCm과 SepalWidthCm을 색상을 달리해 표현

plt.scatter(df.loc[df["Species"] == "Iris-setosa", "SepalLengthCm"], df.loc[df["Species"] == "Iris-setosa", "SepalWidthCm"], color="r")
plt.scatter(df.loc[df["Species"] == "Iris-versicolor", "SepalLengthCm"], df.loc[df["Species"] == "Iris-versicolor", "SepalWidthCm"], color="g")
plt.scatter(df.loc[df["Species"] == "Iris-virginica", "SepalLengthCm"], df.loc[df["Species"] == "Iris-virginica", "SepalWidthCm"], color="b")

plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title("iris")
plt.show()
```

```python
# Species에 따라 SepalLengthCm과 SepalWidthCm을 색상을 달리해 표현

plt.scatter(df.loc[df["Species"] == "Iris-setosa", "PetalLengthCm"], df.loc[df["Species"] == "Iris-setosa", "PetalWidthCm"], color="r")
plt.scatter(df.loc[df["Species"] == "Iris-versicolor", "PetalLengthCm"], df.loc[df["Species"] == "Iris-versicolor", "PetalWidthCm"], color="g")
plt.scatter(df.loc[df["Species"] == "Iris-virginica", "PetalLengthCm"], df.loc[df["Species"] == "Iris-virginica", "PetalWidthCm"], color="b")

plt.xlabel("PetalLengthCm")
plt.ylabel("PetalWidthCm")
plt.title("iris")
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)  
ax2 = fig.add_subplot(1, 2, 2) 

ax1 = plt.scatter(df.loc[df["Species"] == "Iris-setosa", "SepalLengthCm"], df.loc[df["Species"] == "Iris-setosa", "SepalWidthCm"], color="r")
ax1 = plt.scatter(df.loc[df["Species"] == "Iris-versicolor", "SepalLengthCm"], df.loc[df["Species"] == "Iris-versicolor", "SepalWidthCm"], color="g")
ax1 = plt.scatter(df.loc[df["Species"] == "Iris-virginica", "SepalLengthCm"], df.loc[df["Species"] == "Iris-virginica", "SepalWidthCm"], color="b")
ax1 = plt.xlabel("SepalLengthCm")
ax1 = plt.ylabel("SepalWidthCm")

ax2 = plt.scatter(df.loc[df["Species"] == "Iris-setosa", "PetalLengthCm"], df.loc[df["Species"] == "Iris-setosa", "PetalWidthCm"], color="r")
ax2 = plt.scatter(df.loc[df["Species"] == "Iris-versicolor", "PetalLengthCm"], df.loc[df["Species"] == "Iris-versicolor", "PetalWidthCm"], color="g")
ax2 = plt.scatter(df.loc[df["Species"] == "Iris-virginica", "PetalLengthCm"], df.loc[df["Species"] == "Iris-virginica", "PetalWidthCm"], color="b")
ax2 = plt.xlabel("PetalLengthCm")
ax2 = plt.ylabel("PetalWidthCm")

plt.show()
```

```python
mydata = [[1, 2, 3, 4], [10, 11, 12, 13], [30, 31, 32, 33], [41, 43, 45, 47]]

fig, ax = plt.subplots(2, 2)
# ax[0, 0].plot(mydata[0])
# ax[0, 1].plot(mydata[1])
# ax[1, 0].plot(mydata[2])
# ax[1, 1].plot(mydata[3])

for i in range(len(mydata) - 2):
    ax[0, i].plot(mydata[i])
    ax[1, i].plot(mydata[i + 2])

plt.show()
```

```python
import numpy as np

img = Image.open("dog.jpg")

arr = np.array(img)

plt.xticks([])
plt.yticks([])
plt.figure(figsize=(5, 5))

plt.imshow(arr)
```

```python
img1 = np.array(Image.open("dog.jpg"))
img2 = np.array(Image.open("cat.jpg"))
img3 = np.array(Image.open("rabbit.png"))

img_arr = np.array([img1, img2, img3])

fig, ax = plt.subplots(1, 3)

for i in range(len(img_arr)):
    ax[i].plot(img_arr[i])
    
plt.show()
```

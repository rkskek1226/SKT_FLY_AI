```python
df["Activity"] = df["Activity"].map({"LAYING": 0, "STANDING": 1, "SITTING": 2, "WALKING": 3, "WALKING_UPSTAIRS": 4, "WALKING_DOWNSTAIRS": 5})
df
```

```python
x_data = df.drop(["Activity"], axis=1)
x_data
```

```python
y_data = df["Activity"]
y_data
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=1)
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(x_train)
x_train = ss.transform(x_train)
x_test = ss.transform(x_test)
y_train = y_train.values
y_test = y_test.values
```

```python
param_range = [0.001, 0.01, 0.1, 1, 10, 100, 100]
params = [
    {
        "C": param_range,
        "gamma": param_range,
        "kernel": ["rbf"]
    },
    {
        "C": param_range,
        "kernel": ["linear"]
    },
    {
        "C": param_range,
        "kernel": ["poly"],
        "degree": [2, 3]
    }
]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC
from sklearn.svm import SVC

clf = SVC(random_state=1)
gs = GridSearchCV(estimator=clf, param_grid=params, scoring="accuracy", cv=3, n_jobs=-1, verbose=3)
gs.fit(x_train, y_train)
```
